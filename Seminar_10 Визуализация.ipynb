 ## `set_theme`

 https://seaborn.pydata.org/generated/seaborn.set_theme.html

```python
sns.set_theme(style="whitegrid")
```

Этот код устанавливает тему (theme) для графиков, которая определяет стиль и внешний вид графиков, создаваемых с помощью Seaborn.

В данном случае, `style="whitegrid"` указывает на использование темы `"whitegrid"`. Тема `"whitegrid"` задает белый фон для графиков и добавляет сетку (grid) на фоне, что облегчает восприятие данных на графиках и делает их более читаемыми.


* `"darkgrid"`: Это тема с темным фоном и сеткой на фоне для белых графических элементов.

* `"whitegrid"`: Эта тема имеет белый фон и сетку на фоне для черных графических элементов.

* `"dark"`: Тема с темным фоном без сетки.

* `"white"`: Тема с белым фоном без сетки.

* `"ticks"`: Эта тема использует темный фон и метки (текст) с черными осями.

````python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=df, x='households', y='population', hue='total_rooms')
````

## Параметр `palette` в `scatterplot`

Параметр `palette` определяет цветовую палитру, которая будет использоваться для кодировки данных на графике.

Палитра `"flare"` является одной из предопределенных цветовых палитр в библиотеке Seaborn. Эта палитра предоставляет набор ярких и насыщенных цветов, что может быть полезно для выделения данных на графике.

````python
sns.scatterplot(data=df,
                x='households',
                y='population',
                hue='total_rooms',
                palette="flare")
````

Вы можете выбирать разные цветовые палитры в зависимости от ваших предпочтений и требований вашего проекта. Seaborn предоставляет множество предопределенных палитр, и вы также можете создавать собственные палитры, чтобы лучше соответствовать вашим потребностям в визуализации данных.

http://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette

* Палитра `"deep"`: Эта палитра предоставляет яркие и насыщенные цвета, которые могут быть полезны для выделения данных на графиках. Она хорошо подходит для графиков, где нужно выделить несколько категорий.


* Палитра `"muted"` также предоставляет приглушенные и пастельные цвета. Она может быть полезна для создания более спокойных и менее контрастных графиков.

* Палитра `"pastel"`: Эта палитра содержит цвета, которые напоминают пастельные оттенки. Она создает мягкий и нежный внешний вид для графиков.

* Палитра `"dark"` использует более темные и насыщенные цвета. Она может быть полезна для создания графиков с более выраженными контрастами.
"colorblind":

* Палитра `"colorblind"` оптимизирована для доступности и учитывает цветовосприятие людей с дефицитом цветового зрения. Она обеспечивает хороший контраст между цветами.

## Задаём размер через `plt.figure()` из `Matplotlib`
Для увеличения размера графика в `Seaborn`, вы можете использовать функцию `plt.figure()` из библиотеки `Matplotlib`, которая предоставляет возможность настройки размеров графика до создания непосредственно самого графика.

````python
# Создание новой фигуры с установкой размеров
plt.figure(figsize=(10, 6))  # Указываем ширину и высоту фигуры

# Создание scatterplot с настройками
sns.scatterplot(data=df,
                x='households',
                y='population',
                hue='total_rooms',
                palette="crest",
                size='housing_median_age')

# Опционально добавление легенды
plt.legend(title='Total Rooms')

# Отображение графика
plt.show()
````

Использование функции `plt.figure(figsize=(width, height))` из библиотеки `Matplotlib` для настройки размеров графика может показаться необычным, но это объясняется тем, что Seaborn внутренне использует библиотеку `Matplotlib` для создания и настройки графиков. `Matplotlib` - это более низкоуровневая библиотека для создания графиков в Python, и Seaborn является высокоуровневой библиотекой, которая предоставляет удобные методы для создания красивых статистических графиков.

Иногда Seaborn может не предоставлять непосредственного доступа к некоторым функциям `Matplotlib`, поэтому для выполнения определенных настроек (например, изменения размера графика) приходится использовать `Matplotlib` напрямую. Это обеспечивает гибкость в настройке графиков, когда требуется больше управления, чем предоставляет `Seaborn`.

Таким образом, использование `plt.figure(figsize=(width, height))` - это простой и эффективный способ настроить размеры графика, когда вы работаете с `Seaborn`. Вы можете комбинировать возможности обеих библиотек, чтобы создавать красочные и информативные графики с настраиваемыми параметрами.

### Задаём размер через `sns.set` из `Seaborn`

 Seaborn, для настройки размера графика, вы можете использовать функцию
 ```python
 seaborn.set(rc={'figure.figsize': (width, height)})
 ```
 где `width` и `height` - это ширина и высота графика в дюймах соответственно. Эта функция устанавливает параметры рисунка (`figure`) в Seaborn и будет влиять на все последующие графики, созданные с помощью Seaborn.



````python
# Установка размера графика для всех последующих графиков
sns.set(rc={'figure.figsize': (10, 6)})

# Создание scatterplot с установленным размером
sns.scatterplot(data=df,
                x='households',
                y='population',
                hue='total_rooms',
                palette="crest",
                size='housing_median_age')

# Опционально добавление легенды
plt.legend(title='Total Rooms')

# Отображение графика
plt.show()
````

## Заголовки и подписи осей

* `plt.title("Scatterplot of Households vs. Population")` добавляет заголовок к графику.

* `plt.xlabel("Households")` и `plt.ylabel("Population")` добавляют подписи к осям x и y соответственно.

Используя функции Matplotlib, вы можете настраивать текст, размер шрифта, положение и другие параметры заголовков и подписей осей по вашему усмотрению.

````python
# Создание scatterplot
sns.scatterplot(data=df,
                x='households',
                y='population',
                hue='total_rooms',
                palette="crest",
                size='housing_median_age')

# Добавление заголовка к графику
plt.title("Scatterplot of Households vs. Population")

# Добавление подписей осей
plt.xlabel("Households")
plt.ylabel("Population")

# Опционально добавление легенды
plt.legend(title='Total Rooms')

# Отображение графика
plt.show()
````

### Использование `fontdict` для подписей осей и заголовков

https://matplotlib.org/stable/gallery/text_labels_and_annotations/font_family_rc.html

````python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'sans-serif'}
lable_dict = {'fontsize': '14', 'color': '#808080', 'family': 'sans-serif'}
````

````python
# Создание scatterplot
sns.scatterplot(data=df, x='households', y='population', hue='total_rooms', palette="crest", size='housing_median_age')

# Добавление заголовка к графику
plt.title("Scatterplot of Households vs. Population", fontdict=title_dict)

# Добавление подписей осей
plt.xlabel("Households", fontdict=lable_dict)
plt.ylabel("Population", fontdict=lable_dict)

# Опционально добавление легенды
plt.legend(title='Total Rooms')

# Отображение графика
plt.show()
````

## Добавим новую группу

Разобьем мединный возраст домов на 3 категории: новые дома - 20 лет, от 20 до 50 и от 50, чтобы посмотреть влияет ли это на доход.

````python
df.loc[df['housing_median_age'] <= 20, 'age_houseing_group'] = 'Новые'
df.loc[(df['housing_median_age'] > 20) & (df['housing_median_age'] <= 50), 'age_houseing_group'] = 'Средние'
df.loc[df['housing_median_age'] > 50, 'age_houseing_group'] = 'Старые'
````

`median_house_value` - Медианная стоимость дома для семей в рамках квартала (измеряется в долларах США)

`housing_median_age` - Средний возраст дома в квартале; меньшее число означает более новое здание.



````python
sns.histplot(data=df, x="median_house_value", hue="age_houseing_group", palette="husl")

plt.title("Распределение стоимости жилья в зависимости от возраста", fontdict=title_dict)
plt.xlabel("Стоимость жилья в тысяч долларов США", fontdict=lable_dict)
plt.ylabel("Количество", fontdict=lable_dict)

plt.show()
````

# Линейные графики или же `relplot`


````python
df.head(1)
````

`median_house_value` - Медианная стоимость дома для семей в рамках квартала (измеряется в долларах США)

`housing_median_age` - Средний возраст дома в квартале; меньшее число означает более новое здание.



````python
# plt.figure(figsize=(10, 6))  # Указываем ширину и высоту фигуры
sns.relplot(data=df, x="housing_median_age", y="median_house_value", height=6, aspect=1.67)

plt.title("Распределение зарплат", fontdict=title_dict)
plt.xlabel("Средний возраст дома в квартале", fontdict=lable_dict)
plt.ylabel("Медианная стоимость дома", fontdict=lable_dict)

plt.show()
````

## Параметр `kind`в функции `sns.relplot`

Параметр `kind` в функции `sns.relplot` определяет форму и структуру графика, который вы хотите построить на основе ваших данных. В данном случае `kind="line"` означает, что будет построен линейный график.

* `"scatter"`: Точечный график (scatter plot) - это график, на котором данные представлены точками, а каждая точка соответствует одной паре значений (x, y).

* `"line"`: Линейный график (line plot) - это график, на котором данные представлены линиями, соединяющими точки данных в порядке их следования по оси x.

````python
# plt.figure(figsize=(10, 6))  # Указываем ширину и высоту фигуры

# Для визуализации выбирается тип line
sns.relplot(data=df, x="housing_median_age", y="median_house_value", kind="line", height=6, aspect=1.67)

plt.title("Влияние возраста на стоимость жилья", fontdict=title_dict)
plt.xlabel("Средний возраст дома в квартале", fontdict=lable_dict)
plt.ylabel("Медианная стоимость дома", fontdict=lable_dict)

plt.show()
````

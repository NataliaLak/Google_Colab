 ## `set_theme`

 https://seaborn.pydata.org/generated/seaborn.set_theme.html

```python
sns.set_theme(style="whitegrid")
```

Этот код устанавливает тему (theme) для графиков, которая определяет стиль и внешний вид графиков, создаваемых с помощью Seaborn.

В данном случае, `style="whitegrid"` указывает на использование темы `"whitegrid"`. Тема `"whitegrid"` задает белый фон для графиков и добавляет сетку (grid) на фоне, что облегчает восприятие данных на графиках и делает их более читаемыми.


* `"darkgrid"`: Это тема с темным фоном и сеткой на фоне для белых графических элементов.

* `"whitegrid"`: Эта тема имеет белый фон и сетку на фоне для черных графических элементов.

* `"dark"`: Тема с темным фоном без сетки.

* `"white"`: Тема с белым фоном без сетки.

* `"ticks"`: Эта тема использует темный фон и метки (текст) с черными осями.

````python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(6, 4))
plt.hist(df['Memory_Amount'], bins=10)
plt.title('Анализ ноутбуков в зависимости от объема памяти')
plt.xlabel('Объем памяти, GB')
plt.ylabel('Количесво')
````

````python
sns.set_theme(style="darkgrid")

plt.figure(figsize=(6, 4))
plt.hist(df['Memory_Amount'], bins=10)
plt.title('Анализ ноутбуков в зависимости от объема памяти')
plt.xlabel('Объем памяти, GB')
plt.ylabel('Количесво')
````

## `log_scale=True`

Параметр `log_scale=True` используется для установки логарифмической шкалы для одной или обеих осей графика. Когда вы устанавливаете `log_scale=True`, данные на соответствующей оси (обычно y-оси) будут отображаться в логарифмическом масштабе.

Логарифмический масштаб полезен, когда у вас есть данные, которые охватывают большие диапазоны значений. Он помогает лучше визуализировать и анализировать данные, особенно если есть значительные различия в масштабах между разными точками данных.

### Планеты
Давайте рассмотрим пример с данными об относительных размерах планет в нашей Солнечной системе. Эти данные обычно охватывают широкий диапазон значений, что делает логарифмическую шкалу полезной. Сформируем логарифмическую гистограмму:

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Создание DataFrame с размерами планет в км
planet_sizes = pd.DataFrame({
    'Планета': ['Меркурий', 'Венера', 'Земля', 'Марс', 'Юпитер', 'Сатурн', 'Уран', 'Нептун'],
    'Размер (км)': [4879, 12104, 12742, 6779, 139820, 116460, 50724, 49244]
})

# Создание графика с логарифмической шкалой на y-оси
plt.figure(figsize=(10, 6))
sns.barplot(x='Планета', y='Размер (км)', data=planet_sizes, log=True, color='skyblue')
plt.title('Размеры планет в Солнечной системе (логарифмическая шкала)')
plt.xlabel('Планета')
plt.ylabel('Размер (км)')

# Отображение графика
plt.show()
````

3.1 Постройте график в абсолютных величинах

````python
import numpy as np
````

````python
# Готовим данные для графика
data = pd.crosstab(df['Company'], df['Cpu_Company']).reset_index()
data.rename(columns={0: 'Intel', 1: 'AMD'}, inplace=True)
data.head()
````

````python
# Строим столбчатую многорядную диаграмму

plt.figure(figsize=(10, 8))

# определяем кол-во делений
n_ticks = np.arange(len(data['Intel']))

# определяем сдвиг
offset = 0.2

# определяем ширину столбцов
w = 0.4

# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks - offset, data['Intel'], width=w)
plt.bar(n_ticks + offset, data['AMD'], width=w)

plt.title('Взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров')
plt.xlabel('Производители ноутбуков')
plt.ylabel('Кол-во')
plt.legend(['Intel', 'AMD'])
plt.xticks(n_ticks, data['Company'], rotation = 30);  # добавляем метки делений
````

````python
df_company_process = pd.crosstab(df['Company'], df['Cpu_Company'])
df_company_process.drop('Samsung', axis=1, inplace = True)
plt.bar(df_company_process.index,df_company_process['AMD'], color='green')
plt.bar(df_company_process.index,df_company_process['Intel'], bottom = df_company_process['AMD'])

plt.xticks(rotation = 45);
````

3.2 Постройте график в относительных величинах

````python
data = pd.crosstab(df['Company'], df['Cpu_Company'], normalize='index').reset_index() ##.sort_values(by=0, ascending=False)
data.rename(columns={0: 'AMD', 1: 'Intel'}, inplace=True)
plt.figure(figsize=(15, 6))

plt.bar(data['Company'], data['AMD'])
plt.bar(data['Company'], data['Intel'], bottom=data['AMD'])

plt.title('Производители процессоров')
plt.xlabel('призводители ноутбуков')
plt.ylabel('Доля процессоров')
plt.legend(['AMD', 'Intel'])
plt.xticks(rotation = 30);
````
